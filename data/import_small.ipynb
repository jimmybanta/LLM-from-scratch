```python
import os
import json
import unicodedata
from unidecode import unidecode
```

```python
dataset_path = os.environ['SMALL_DATASET_PATH']
```

# Read the data

```python
with open(dataset_path) as f:
    data = json.load(f)
```

```python
# Data is a list of dictionaries - where each dictionary represents a text exchange between a human and an AI.
# Each dictionary contains an ID, and a 'conversations' key.
# The 'conversations' key contains a list of dictionaries, where each dictionary contains the author (either 'human' or 'gpt'), and the text that they said.
```

```python
# all_text will be a massive string that contains all of the text in the dataset.
all_text = ''

for i, conversation in enumerate(data):
    for item in conversation['conversations']:
        all_text += item['value'] + ' '
    print(f'Processed conversation {i+1}')
```

```python
# get a rough word count
total_words = len(all_text.split())
```
